```python
import numpy as np
import pandas as pd
```

```python
df_film_30ans = pd.read_csv("df_film_30ans_image_syn_ml.csv")
df_kids = pd.read_csv("df_kids.csv")
```

```python
df_film_30ans
```

```python
df_kids
```

```python
lst_tconst_30ans = list(df_film_30ans['tconst'])
lst_title_30ans = list(df_film_30ans['title'])
lst_tconst_kids = list(df_kids['tconst'])
lst_title_kids = list(df_kids['title'])
```

```python
# 30 ans
lst_options_films =[]
for i in range (len(lst_tconst_30ans)) :
  dic_film = {'label' : '', 'value' : ''}
  dic_film['label']= lst_title_30ans[i]
  dic_film['value'] = lst_tconst_30ans[i]
  lst_options_films.append(dic_film)
```

```python
# 30 ans + kids
for i in range (len(lst_tconst_kids)): 
  dic_film = {'label' : '', 'value' : ''}
  dic_film['label']= lst_title_kids[i]
  dic_film['value'] = lst_tconst_kids[i]
  lst_options_films.append(dic_film)
```
